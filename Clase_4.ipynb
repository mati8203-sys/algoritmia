METODO GROUPBY

```python
import pandas as pd

from datetime import datetime

facturas = pd.DataFrame({'folio':(15234,

           15235,

           15236,

           15237,

           15238,

           15239,

           15240,

           15241,

           15242),

      'sucursal':('CDMX01',

            'MTY01',

            'CDMX02',

            'CDMX02',

            'MTY01',

            'GDL01',

            'CDMX02',

            'MTY01',

            'GDL01'),

      'monto':(1420.00,

          1532.00,

          890.00,

          1300.00,

          3121.47,

          1100.5,

          12230,

          230.85,

          1569),

      'fecha':(datetime(2019,3,11,17,24),

          datetime(2019,3,24,14,46),

          datetime(2019,3,25,17,58),

          datetime(2019,3,27,13,11),

          datetime(2019,3,31,10,25),

          datetime(2019,4,1,18,32),

          datetime(2019,4,3,11,43),

          datetime(2019,4,4,16,55),

          datetime(2019,4,5,12,59)),

     'cliente':(19234,

           19232,

           19235,

           19233,

           19236,

           19237,

           19232,

           19233,

           19232)

})
```

```python
gastosporcliente= facturas.filter(items=["monto","cliente"])
print(gastosporcliente)
```

```python
clientes = gastosporcliente.groupby("cliente")
print(clientes.groups)
```

```python
for nombre_cliente , grupo in clientes:
  print(f"Grupo para cliente: (nombre_clientes)")
```
