ARREGLOS

```python
import numpy as np
arreglo = np.array([2,2,2,2], dtype = float)
print(arreglo)
```

DIMENSIONES de los arreglos

```python
print("el numero de dimensiones del arreglo es : ", arreglo.ndim)
```

ARREGLO.size = numero de elementos
arreglo.shape = forma del arreglo
arreglo.dtype = tipo de datos(float,int, input) etc

EJEMPLO

```python
arreglo3d = np.array([
    [[1,2,3],[5,6,7]],
    [[7,8,9],[10,11,12]]
])
```

```python
print("el numero de dimensiones del arreglo es : ", arreglo3d.ndim)
print("la forma del arreglo es:", arreglo3d.shape)
print("el numero total de elementos del arreglo es:", arreglo3d.size)
print("los tipos de datos del arreglo son:", arreglo3d.dtype)
```
