Analizar cuales son las columnas de mi interes

```python
data.columns
```

```python
data.drop(["zipcode", "lat","long"], axis = 1 , inplace =True)
```

```python
data.shape
```

```python
data['house_age'] = 2026 - data['yr_built']
```

```python
data['house_age']
```

```python
data['renovated_flag'] = data['yr_renovated'].apply(lambda x: 1 if x > 0 else 0)
```

```python
data[(data['bedrooms'] ==3) & (data['floors'] > 7)]
```

```python
df_line = data.groupby('bedrooms') [ 'price'].mean()
plt.plot(df_line.index, df_line.values)
plt.title("Precio promedio según número de habitaciones")
plt.xlabel("Número de habitaciones")
plt.ylabel("Precio promedio")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
data['bedrooms'].value_counts().plot(kind='bar')
plt.title("Distribución de número de habitaciones")
plt.xlabel("Habitaciones")
plt.ylabel("Cantidad de viviendas")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(data['price'], bins=50)
plt.title("Histograma de precios de vivienda")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.tight_layout()
plt.show()
```
